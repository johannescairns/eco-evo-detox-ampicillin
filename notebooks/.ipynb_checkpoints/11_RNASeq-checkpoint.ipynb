# Analysis of RNASeq DE data

```r
# See: 
# https://www.bioconductor.org/packages/release/bioc/vignettes/DESeq2/inst/doc/DESeq2.html
# https://www.bioconductor.org/packages/release/workflows/vignettes/rnaseqGene/inst/doc/rnaseqGene.html
# https://www.youtube.com/watch?v=ht1r34-ifVI
```

```r
# Read in data

comms = read.table("../data/comparing_communities_with_abundance_info.tsv", header = T, sep = "\t")
pulses = read.table("../data/comparing_treatments_with_abundance_info.tsv", header = T, sep = "\t")
head(comms)
head(pulses)
```

```r
unique(comms$treatment)
unique(comms$comparison)
unique(pulses$community)
unique(pulses$comparison)
#a. ANC (all ancestral strains)
#b. E1287 (all ancestral expect evolved 1287)
#c. E1896 (all ancestral except evolved 1896)
#d. E1972 (all ancestral except evolved 1972)
#e. E1977 (all ancestral except evolved 1977)
#f. EVO (all evolved strains)
#Treatment 1. 5 ug/ml (t1)
#Treatment 2. 50 ug/ml (t2)
#Treatment 3. 500 ug/ml (t3)
#Treatment 4. (control) 0 ug/ml (t4)
```

```r
# Rename "communities" and "treatments"

comms$treatment = factor(comms$treatment, 
                         levels = c('t4','t1','t2','t3'), 
                         labels = c(0,5,50,500))
comms$comparison = factor(comms$comparison, 
                         levels = c('a vs. d','a vs. e','a vs. f','d vs. e','d vs. f','e vs. f'),
                         labels = c('ANC vs. E1972',
                                    'ANC vs. E1977',
                                    'ANC vs. EVO',
                                    'E1972 vs. E1977',
                                    'E1972 vs. EVO',
                                    'E1977 vs. EVO'))
pulses$community = factor(pulses$community, levels = c('a','d','e'),
                         labels = c("ANC", "E1972", "E1977"))
pulses$comparison = factor(pulses$comparison, 
                          levels = c('t1 vs. t2','t1 vs. t3','t1 vs. t4','t2 vs. t3','t2 vs. t4','t3 vs. t4'),
                         labels = c('5 vs. 50','5 vs. 500','5 vs. 0','50 vs. 500','50 vs. 0','500 vs. 0'))
```

```r
# Load dependencies

library(reshape2)
library(tidyverse)
```

```r
#head(comms[,-c(6:8)])
#head(pulses[,-c(6:8)])

# Keep only genes where both strains at min 1 % relative abundance

#max(comms$ab1)
#max(comms$ab2)
#max(pulses$ab1, na.rm = T)
#max(pulses$ab2, na.rm = T)

comms$sufficient_ab = ifelse(comms$ab1 >= 0.01 & comms$ab2 >= 0.01, "ok", "bad")
pulses$sufficient_ab = ifelse(pulses$ab1 >= 0.01 & pulses$ab2 >= 0.01, "ok", "bad")

comms_suff = comms[comms$sufficient_ab == "ok",]
pulses_suff = pulses[pulses$sufficient_ab == "ok",]

nrow(comms)
nrow(comms_suff)
nrow(pulses)
nrow(pulses_suff)

head(comms_suff[,-c(6:8)])
head(pulses_suff[,-c(6:8)])
```

```r
# Remove cases where species abundance change and gene expression change go in same direction and
# gene expression change less than 10-fold of abundance change

comms_suff$ab_log2fold = log2(comms_suff$ab1)-log2(comms_suff$ab2)
pulses_suff$ab_log2fold = log2(pulses_suff$ab1)-log2(pulses_suff$ab2)
#head(comms_suff[,-c(1:3,6:8)])
#head(pulses_suff[,-c(1:3,6:8)])

comms_suff$same_direction = NA
comms_suff$same_direction = ifelse(comms_suff$log2FoldChange < 0 & 
                                   comms_suff$ab_log2fold < 0 & 
                                   (comms_suff$log2FoldChange*(-1) < (10*comms_suff$ab_log2fold*(-1))),
                                  "bad", comms_suff$same_direction)
comms_suff$same_direction = ifelse(comms_suff$log2FoldChange > 0 & 
                                   comms_suff$ab_log2fold > 0 & 
                                   (comms_suff$log2FoldChange < (10*comms_suff$ab_log2fold)),
                                  "bad", comms_suff$same_direction)
comms_suff$same_direction = ifelse(is.na(comms_suff$same_direction), "ok", comms_suff$same_direction)
comms_ok = comms_suff[comms_suff$same_direction == "ok",]
head(comms_ok[,-c(1:3,6:12)])
comms_ok[,-c(1:3,6:12)]
nrow(comms_ok)
```

```r
# Remove cases where species abundance change and gene expression change go in same direction and
# gene expression change less than 10-fold of abundance change

pulses_suff$ab_log2fold = log2(pulses_suff$ab1)-log2(pulses_suff$ab2)
pulses_suff$ab_log2fold = log2(pulses_suff$ab1)-log2(pulses_suff$ab2)
#head(pulses_suff[,-c(1:3,6:8)])
#head(pulses_suff[,-c(1:3,6:8)])

pulses_suff$same_direction = NA
pulses_suff$same_direction = ifelse(pulses_suff$log2FoldChange < 0 & 
                                   pulses_suff$ab_log2fold < 0 & 
                                   (pulses_suff$log2FoldChange*(-1) < (10*pulses_suff$ab_log2fold*(-1))),
                                  "bad", pulses_suff$same_direction)
pulses_suff$same_direction = ifelse(pulses_suff$log2FoldChange > 0 & 
                                   pulses_suff$ab_log2fold > 0 & 
                                   (pulses_suff$log2FoldChange < (10*pulses_suff$ab_log2fold)),
                                  "bad", pulses_suff$same_direction)
pulses_suff$same_direction = ifelse(is.na(pulses_suff$same_direction), "ok", pulses_suff$same_direction)
pulses_ok = pulses_suff[pulses_suff$same_direction == "ok",]
head(pulses_ok[,-c(1:3,6:12)])
pulses_ok[,-c(1:3,6:12)]
nrow(pulses_ok)
```
