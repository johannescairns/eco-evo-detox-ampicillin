```r
df = read.table("../data/AMP_growth_k_auc.txt", header = T, check.names = F, sep = "\t")
head(df)
```

```r
# Use medium control data to infer threshold level for no growth

#unique(df$Species)
#df[df$Species == "only medium",]
max(df$k[df$Species == "only medium"])
mean(df$k[df$Species == "only medium"])

# Inspect mean and max levels in the absence of antibiotics

aggregate(data = df[df$AB == 0,], k ~ Species, FUN = mean) # Species with lowest k = 0.14
aggregate(data = df[df$AB == 0,], k ~ Species, FUN = max) # Species with lowest k = 0.14

# Any value above 1.3 seems to result from no growth (low strain growth causing unreliable k estimation)
# Any value below 0.1 seems to be unreliable

# Let's convert these values to 0

df$k = ifelse(df$k < 0.1 | df$k > 1.3, 0, df$k)
max(df$k)
min(df$k)
```

```r
# Rename strains

df_glob = df %>%
  mutate(strainID = paste0("HAMBI-", str_pad(Species, 4, pad = c("0"))))
head(df_glob)
```

```r
# Define treatments
df_glob2$Treat = paste(df_glob2$Pop, df_glob2$Species, df_glob2$Replicate, sep = "_")
head(df_glob2)
Treatments = unique(df_glob2$Treat)
Treatments
```

```r
# Get taxon names

# Read data ---------------------------------------------------------------
counts <- read_tsv("../data/species_counts_correct.tsv")
```

```r
# Remove control treatment

nrow(df_glob)
df_glob = df_glob[df_glob$Species != "only medium",]
nrow(df_glob)

# Define treatments
df_glob$Treat = paste(df_glob$Pop, df_glob$strainID, df_glob$Replicate, sep = "_")
head(df_glob)
Treatments = unique(df_glob$Treat)
#Treatments
```

```r
# For each species, replicate and anc vs evo, compute MIC as the highest concentration where k > 0
# omitting positive values after two successive zero counts as errors

# Convert read count data to proportions and modify column names

df_glob$MIC = NA
for(i in 1:length(Treatments)){
    a = df_glob[df_glob$Treat == Treatments[i],]
    a = a[order(a$AB),]
    a$nextzero = NA
    for(j in 1:nrow(a)){
        a$nextzero[j] = ifelse(a$k[j] == 0 & a$k[j+1] == 0, "yes", "no")
    }
    minnextzero = ifelse(length(a$AB[a$nextzero == "yes"]) == 0, 2050, min(a$AB[a$nextzero == "yes"], na.rm = T))
    a = a[a$AB<minnextzero,]
    a_MIC = max(a$AB, na.rm = T)
    a_MIC[is.infinite(a_MIC)] <- 0
    df_glob$MIC[df_glob$Treat == Treatments[i]] = a_MIC
}

head(df_glob)
```

```r
# Keep only MIC per condition

df_glob2 = df_glob[!duplicated(df_glob$Treat),c("strainID", "Pop", "Species", "Replicate", "Treat", "MIC")]
nrow(df_glob)
nrow(df_glob2)
head(df_glob2)
df_glob2[df_glob2$MIC > 1900,] # In each case of max MIC at least one pop able to grow above highest
# concentration --> 2000 is minimum MIC for these strains
```

```r
library(tidyverse)
library(patchwork)
library(here)
library(broom)
library(withr)
library(ggforce)
library(scales)

source("utils_IC50.R")
```

```r
progenitor = progenitor %>%
  unite(id, strainID, replicate, sep="_", remove=FALSE) %>%
  arrange(id)
```

```r
# format data -------------------------------------------------------------

# generate names vector for lists.
my_names <- progenitor %>%
  distinct(id) %>%
  arrange(id) %>%
  pull(id)

# split data frame by treatment and id into list
my_split <- progenitor %>% 
  arrange(id) %>%
  group_by(id) %>%
  arrange(desc(ampicillin)) %>%
  group_split()

names(my_split) <- my_names
```

```r
# nls fit 3 and 4 paramater log-logistic to data --------------------------

# 4 param model
with_seed(
  1234,
  my_split_fit_4p <-
    map(
      my_split,
      safe_ll_fit,
      OD = OD600,
      AB = amplicillin,
      model = m4pll,
      model_start = m4pll_start
    )
)

# 3 param model
with_seed(
  1234,
  my_split_fit_3p <-
    map(
      my_split,
      safe_ll_fit,
      OD = OD600,
      AB = ampicillin,
      model = m3pll,
      model_start = m3pll_start
    )
)

names(my_split_fit_4p) <- my_names
names(my_split_fit_3p) <- my_names
```

```r
unique(round(rev(my_split[[1]]$ampicillin)))
```

```r
# 4 parameter models failed
```

```r
# extract the IC50 values from model fits ---------------------------------

# the 'e' parameter is the IC50 from the 4 and 3 parameter log-logistic models.
# Even though the log-logistic will fit a dose response curve that never
# goes to 0, we don't trust model fits from OD600 that did not decrease
# below 0.1 because these strains didn't die. They may have decreased, but not
# enough to be considered genuinely inhibited by the antibiotic. 
# Usually when OD600 is always > 0.1 it means the Dose-response curve is 
# just a horizontal line which makes it impossible to fit a meaningful 
# log-logistic model.

# find minimum OD value. This is used to identify clones that 
my_mins <- progenitor %>%
  group_by(id) %>%
  summarize(min=min(OD600)) %>%
  ungroup() %>%
  arrange(id)

# get tidied model results
my_IC50x_3p <- map_df(my_split_fit_3p, tidy, .id="id") %>%
  pivot_wider(id_cols=c(id), names_from=term, values_from=estimate) %>%
  dplyr::select(id, IC50_3p=e)

my_IC50x <- my_IC50x_3p %>%
  left_join(., my_mins) %>%
  pivot_longer(cols = IC50_3p, names_to = "IC50_mod", values_to = "IC50x") %>%
  group_by(id) %>%
  mutate(IC50_mod = case_when(is.na(unique(IC50x)) ~ NA_character_,
                              TRUE ~ IC50_mod)) %>%
  ungroup() %>%
  # if the minimum value of OD600 never goes below 0.1 then 
  # we assume that the clone is completely resistant rather than
  # use model fits. This ensures that we don't get extreme IC50x 
  # values
  mutate(IC50x=ifelse(min > 0.2, 2000, IC50x)) %>%
  filter(IC50x <= 2000 | is.na(IC50x)) %>%
  arrange(id, desc(IC50_mod)) %>%
  group_by(id) %>%
  slice(1) %>%
  ungroup() %>%
  dplyr::select(id, IC50_mod, IC50x) %>%
  separate(id, into=c("strainID", "replicate"), 
           sep="_", remove = FALSE)
```

```r
# save the IC50 values for later
write.table(my_IC50x, file = "../data/result_IC50_init_progenitor.txt", quote = F, sep = "\t")
```

```r
# Check if IC50 has improved for each species

# Since only two (or one if one replicate failed) measurements
# for each condition, this cannot be statistically tested

# Lets take average of two replicates and look at fold change

IC50_means = aggregate(IC50x ~ strainID*evo_history*name, FUN = mean, data = IC50)

# Add anc and evo as rows and take difference

IC50_means_anc = IC50_means[IC50_means$evo_history == "Ancestral species",]
colnames(IC50_means_anc)[4] = "IC50x_anc"
IC50_means_evo = IC50_means[IC50_means$evo_history != "Ancestral species",]
colnames(IC50_means_evo)[4] = "IC50x_evo"

IC50_means2 = merge(IC50_means_anc[,-2], IC50_means_evo[,-2],
                   by = c("strainID", "name"))
IC50_means2$IC50_evo_vs_anc = IC50_means2$IC50x_evo / IC50_means2$IC50x_anc
IC50_means2 = IC50_means2[order(IC50_means2$IC50_evo_vs_anc),]
IC50_means2$strainID = factor(IC50_means2$strainID, levels = IC50_means2$strainID)
head(IC50_means2)
```

```r
# Take species with 2-fold or above difference

IC50_means2$strainID[IC50_means2$IC50_evo_vs_anc <= 0.5 |
                     IC50_means2$IC50_evo_vs_anc >= 2]
```
