# Genomic data visualization and analysis

```r
# Dependencies

#install.packages(c("here", "Polychrome", "withr", "patchwork"), dep = T)
library(here)
library(tidyverse)
library(Polychrome)
library(withr)
library(patchwork)
library(scales)
library(svglite)
```

```r
df_metagen = read_tsv("../data/APEX_all_variant_results_min_coverage_5.tsv", show_col_types = FALSE)
df_wgs = read_tsv("../data/all_WGS_variants_with_genes.tsv", show_col_types = FALSE)
#head(df_metagen)
#head(df_wgs)
```

```r
# Keep relevant columns

df_metagen = df_metagen[,c("Sample","HAMBI", "POS","CHROM", "GENE", "product1", "gene_name1", "AF", "EFFECT", "TYPE", "DP", "HGVS_P")]
df_wgs = df_wgs[,c("Sample", "POS","CHROM", "GENE", "product1", "gene_name1", "AF", "EFFECT", "TYPE", "DP", "HGVS_P")]
```

```r
# Add info on whether same variant is present in evolved strain

df_metagen$uniqvar = paste(df_metagen$CHROM, df_metagen$GENE, df_metagen$HGVS_P, sep = "_")
df_wgs$uniqvar = paste(df_wgs$CHROM, df_wgs$GENE, df_wgs$HGVS_P, sep = "_")
df_metagen$evo_presence = ifelse(df_metagen$uniqvar %in% unique(df_wgs$uniqvar), "yes", "no")
df_wgs$evo_presence = ifelse(df_wgs$uniqvar %in% unique(df_metagen$uniqvar), "yes", "no")
```

```r
# Check number of species in each
unique(df_metagen$HAMBI)
unique(df_wgs$Sample)
```

```r
# Remove all genes with variants in control community except for those also present in pre-exposed clones
# These are likely to be either seq errors or common targets of lab adaptations that are not
# related to the specific experimental conditions

CC = df_metagen[df_metagen$Sample == "CC",]
CC_genes = unique(CC$GENE)
length(CC_genes)
CC_genes = CC_genes[!(CC_genes %in% unique(df_wgs$GENE))]
length(CC_genes)
nrow(df_metagen)
df_metagen = df_metagen[!(df_metagen$GENE %in% CC_genes),]
nrow(df_metagen)
```

```r
# Accept as putative initial variants those in pre-exposed strain WGS data with > 20 read depth and > 0.8 allele frequency
# Remove other genes present in pre-exposed strains also from metagenomic datasets as potential seq errors
# or common targets of lab adaptation unimportant for specific conditions of experiment here

df_wgs2 = df_wgs[df_wgs$DP > 20 & df_wgs$AF > 0.8,] 
evo_gene_to_keep = df_wgs2$GENE
evo_gene_to_remove = df_wgs$GENE[!(df_wgs$GENE %in% evo_gene_to_keep)]
length(evo_gene_to_remove) 
#head(df_wgs2)
length(unique(df_wgs2$GENE)) # 16 genes kept
df_metagen2 = df_metagen[!(df_metagen$GENE %in% evo_gene_to_remove),] # most variants kept
nrow(df_metagen)
nrow(df_metagen2)
```

```r
# Remove spurious variants from metagen dataset and then seek to merge the two datasets

# Keep only those variants that reach a frequency exceeding 0.3 AND depth of 20 at least once 
# OR that are present in pre-exposed strains

maxfreqs = aggregate(data = df_metagen2[df_metagen2$DP>20,], AF ~ uniqvar, FUN = max)
maxfreqs = maxfreqs[order(maxfreqs$AF, decreasing = T),]
#tail(maxfreqs)

#ggplot(maxfreqs, aes(x = AF)) +
#    geom_histogram() 

maxfreq_30_or_more = maxfreqs$uniqvar[maxfreqs$AF>0.3]
length(maxfreq_30_or_more) # keep only 66
nrow(maxfreqs)

df_metagen3 = df_metagen2[df_metagen2$uniqvar %in% maxfreq_30_or_more | df_metagen2$uniqvar %in% unique(df_wgs2$uniqvar),]
nrow(df_metagen2)
nrow(df_metagen3)
length(unique(df_metagen3$uniqvar)) # total 106 variants
```

```r
# Add treatment information

# Evo treatment

df_metagen3$Evo = NA
df_metagen3$Evo = ifelse(grepl("^a-", df_metagen3$Sample), "All ancestral", df_metagen3$Evo)
df_metagen3$Evo = ifelse(grepl("^d-", df_metagen3$Sample), "Pre-exposed 1972", df_metagen3$Evo)
df_metagen3$Evo = ifelse(grepl("^e-", df_metagen3$Sample), "Pre-exposed 1977", df_metagen3$Evo)
df_metagen3$Evo = ifelse(grepl("^f-", df_metagen3$Sample), "All pre-exposed", df_metagen3$Evo)
table(df_metagen3$Evo)

# Replicate

df_metagen3$Replicate = NA
df_metagen3$Replicate = ifelse(grepl("-1-", df_metagen3$Sample), 1, df_metagen3$Replicate)
df_metagen3$Replicate = ifelse(grepl("-2-", df_metagen3$Sample), 2, df_metagen3$Replicate)
df_metagen3$Replicate = ifelse(grepl("-3-", df_metagen3$Sample), 3, df_metagen3$Replicate)
df_metagen3$Replicate = ifelse(grepl("-4-", df_metagen3$Sample), 4, df_metagen3$Replicate)
table(df_metagen3$Replicate)

# Primer pulse treatment

df_metagen3$Primer_pulse = NA
df_metagen3$Primer_pulse = ifelse(grepl("-t1-", df_metagen3$Sample), 5, df_metagen3$Primer_pulse)
df_metagen3$Primer_pulse = ifelse(grepl("-t2-", df_metagen3$Sample), 50, df_metagen3$Primer_pulse)
df_metagen3$Primer_pulse = ifelse(grepl("-t3-", df_metagen3$Sample), 500, df_metagen3$Primer_pulse)
df_metagen3$Primer_pulse = ifelse(grepl("-t4-", df_metagen3$Sample), 0, df_metagen3$Primer_pulse)
table(df_metagen3$Primer_pulse)
```

```r
# Manually inspect those variants absent from metagen data in pre-exposed strain data

df_wgs2[df_wgs2$evo_presence == "no",]
#df_metagen3[df_metagen3$GENE == "H1972_03930",] 
# mutation always present in a position differing by only one nucleotide --> assume bioinfo error in assignment and identify as same variant
df_wgs2$evo_presence[df_wgs2$GENE == "H1972_03930"] = "yes"
df_metagen3$evo_presence[df_metagen3$GENE == "H1972_03930"] = "yes"
#df_metagen2[df_metagen2$GENE == "H1977_01749",] # One case present in one population but mutation located in different position in gene
#df_metagen2[df_metagen2$GENE == "H1977_03891",] # Not present
#df_metagen2[df_metagen2$GENE == "H1977_03892",] # Not present
# The species is driven to very low frequency in the presence of pre-exposure and these variants cannot likely be detected because of that
```

```r
# Require variants present only one community to have over 50x read depth or to be present in pre_exposed strains

df_metagen3$COUNT = 1
var_counts = aggregate(COUNT ~ uniqvar, data = df_metagen3, FUN = sum)
var_counts = var_counts[order(var_counts$COUNT, decreasing = T),]
var_counts_to_keep = var_counts$uniqvar[var_counts$COUNT > 1]
df_metagen4a = df_metagen3[df_metagen3$uniqvar %in% var_counts_to_keep | df_metagen3$evo_presence == "yes",]
nrow(df_metagen3)
nrow(df_metagen4a)
```

```r
df_metagen4b = df_metagen3[!(df_metagen3$uniqvar %in% var_counts_to_keep),]
nrow(df_metagen4b)
df_metagen4b = df_metagen4b[df_metagen4b$DP > 50 | df_metagen4b$evo_presence == "yes",]
nrow(df_metagen4b)
```

```r
df_metagen4 = rbind(df_metagen4a, df_metagen4b)
nrow(df_metagen4)
length(unique(df_metagen4$uniqvar))
```

```r
# Remove variant targets that have more than one independent mutations in the same population 
# (most likely to result from bioinfo/seq errors)

df_metagen4$COUNT = 1
abund_targ = aggregate(data = df_metagen4, COUNT ~ uniqvar + Evo + Primer_pulse + Replicate, FUN = sum)
abund_targ = abund_targ[order(abund_targ$COUNT, decreasing = T),]
#head(abund_targ)

#ggplot(abund_targ, aes(x = COUNT)) +
#    geom_histogram() 

abund_targ_uniqvars = abund_targ$uniqvar[abund_targ$COUNT > 1]
#abund_targ_uniqvars
nrow(df_metagen4)
df_metagen4 = df_metagen4[!(df_metagen4$uniqvar %in% abund_targ_uniqvars),]
nrow(df_metagen4) # 1/3 removed
length(unique(df_metagen4$uniqvar))

#df_metagen4[df_metagen4$uniqvar == "HAMBI_1972_chrm01_circ_H1972_03431-H1972_03432_NA",]

nrow(df_wgs2)
df_wgs2 = df_wgs2[!(df_wgs2$uniqvar %in% abund_targ_uniqvars),]
nrow(df_wgs2) # 18/19 kept
```

```r
# Identify those variant targets present in over 40 populations

df_popanalys = aggregate(data = df_metagen3, AF ~ uniqvar + Evo + Primer_pulse + Replicate, FUN = sum)
df_popanalys$COUNT = 1
df_popanalys2 = aggregate(data = df_popanalys, COUNT ~ uniqvar, FUN = sum)
df_popanalys2 = df_popanalys2[order(df_popanalys2$COUNT, decreasing = T),]
#head(df_popanalys2)

#ggplot(df_popanalys2, aes(x = COUNT)) +
#    geom_histogram() 

# If the same variant is present in over 2/3 of the populations, it is likely to be a common variant
# --> remove those occurring in 40 populations (only three cases)

susp_anc_var = df_popanalys2$uniqvar[df_popanalys2$COUNT>40]
```

```r
df_metagen5 = df_metagen4[!(df_metagen4$uniqvar %in% susp_anc_var),]
nrow(df_metagen4)
nrow(df_metagen5) # reasonable amount, 407 rows left
length(unique(df_metagen5$uniqvar)) # 70 unique variants
nrow(df_wgs2)
df_wgs2 = df_wgs2[!(df_wgs2$uniqvar %in% susp_anc_var),]
nrow(df_wgs2) # 13 variants left
```

```r
# Keep only nonsynonymous mutations

unique(df_metagen5$EFFECT)
df_metagen6 = df_metagen5[!(df_metagen5$EFFECT %in% c('intergenic_region',
                                                    'synonymous_variant',
                                                    'conservative_inframe_insertion',
                                                    'non_coding_transcript_variant',
                                                     'initiator_codon_variant',
                                                     'conservative_inframe_deletion',
                                                     'conservative_inframe_insertion&splice_region_variant')),]
nrow(df_metagen6)
length(unique(df_metagen5$uniqvar)) 
length(unique(df_metagen6$uniqvar)) # 40 variants left

# Keep only nonsynonymous mutations

df_wgs3 = df_wgs2[!(df_wgs2$EFFECT %in% c('intergenic_region',
                                                    'synonymous_variant',
                                                    'conservative_inframe_insertion',
                                                    'non_coding_transcript_variant',
                                                     'initiator_codon_variant',
                                                     'conservative_inframe_deletion',
                                                     'conservative_inframe_insertion&splice_region_variant')),]
nrow(df_wgs3)
length(unique(df_wgs2$uniqvar)) 
length(unique(df_wgs3$uniqvar)) # 10 variants left
```

```r
df_metagen6$Variantgene = paste(df_metagen6$GENE, df_metagen6$gene_name1, sep = "_")
length(unique(df_metagen6$Variantgene))
```

```r
# Aggregate by count

df_metagen6$COUNT = 1
highfreq_aggr = aggregate(COUNT ~ HAMBI*Variantgene*Primer_pulse*Evo*evo_presence*Replicate,FUN = sum, data = df_metagen6)
highfreq_aggr$COUNT = 1
highfreq_aggr2 = aggregate(COUNT ~ HAMBI*Variantgene*Primer_pulse*Evo*evo_presence,FUN = sum, data = highfreq_aggr)
highfreq_aggr2 = highfreq_aggr2[order(highfreq_aggr2$COUNT, decreasing = T),]
max(highfreq_aggr2$COUNT)
min(highfreq_aggr2$COUNT)
#head(highfreq_aggr2)
```

```r
highfreq_aggr2$Evo = factor(highfreq_aggr2$Evo, levels = c("All ancestral", 
                                                         'Pre-exposed 1972',
                                                         'Pre-exposed 1977',
                                                         "All pre-exposed"))
```

```r
H1977 = H1977[!(H1977$ID %in% control_ids_1977),]
nrow(H1977)
```

```r
nrow(H1977)
length(unique(H1977$ID))

# Keep only recurrent targets #same target occurring min 2x in independent treatments
H1977$COUNT = 1
H1977_hits = aggregate(data = H1977, COUNT ~ ID*Community*Treatment*Replicate, FUN = sum)
H1977_hits = H1977_hits[order(H1977_hits$COUNT, decreasing = T),]
head(H1977_hits)
H1977_hits$COUNT = 1
H1977_hits2 = aggregate(data = H1977_hits, COUNT ~ ID, FUN = sum)
H1977_hits2 = H1977_hits2[order(H1977_hits2$COUNT, decreasing = T),]
head(H1977_hits2)
```

```r
p <- ggplot(H1977_hits2, aes(x=COUNT)) + 
  geom_histogram()
p

# use minimum criterion of 3x
```

```r
min3_hits_1977 = H1977_hits2$ID[H1977_hits2$COUNT>2 & H1977_hits2$COUNT<14]

H1977_v2 = H1977[H1977$ID %in% min3_hits_1977,]
nrow(H1977_v2)
H1977_v2$COUNT = 1
```
