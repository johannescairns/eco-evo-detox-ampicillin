```r
# Metagenome data analyses

#install.packages(c("here", "Polychrome", "withr", "patchwork"), dep = T)
library(here)
library(tidyverse)
library(Polychrome)
library(withr)
library(patchwork)
library(scales)
library(svglite)
```

```r
df = read_tsv("../data/APEX_all_variant_results_genenames.tsv")
head(df)
```

```r
# Check locus H1972_00547

df[df$HAMBI == "HAMBI_1972" & df$GENE == "H1972_00547",]
```

```r
colnames(df)
```

```r
sort(unique(df$gene_name1))[1200:1400]
```

```r
# Check locus ftsI in 1972

df[grepl("ftsI", df$gene_name1),-c(4:10)]
```

```r
# Remove mutations in control community from all data

df$mut_pos = paste(df$GENE, df$POS)
length(unique(df$GENE))
length(unique(df$mut_pos))
length(unique(df$GENE[df$Sample == "CC"]))
CC_gene = unique(df$GENE[df$Sample == "CC"])
length(unique(df$mut_pos[df$Sample == "CC"]))
CC_mut = unique(df$mut_pos[df$Sample == "CC"])

nrow(df)
df = df[!(df$mut_pos %in% CC_mut),]
nrow(df) # 17173/33907 mutations retained
```

```r
# Add treatment information

# Evo treatment

df$Evo = NA
df$Evo = ifelse(grepl("^a-", df$Sample), "All ancestral", df$Evo)
df$Evo = ifelse(grepl("^d-", df$Sample), "Evolved 1972", df$Evo)
df$Evo = ifelse(grepl("^e-", df$Sample), "Evolved 1977", df$Evo)
df$Evo = ifelse(grepl("^f-", df$Sample), "All evolved", df$Evo)
table(df$Evo)

# Replicate

df$Replicate = NA
df$Replicate = ifelse(grepl("-1-", df$Sample), 1, df$Replicate)
df$Replicate = ifelse(grepl("-2-", df$Sample), 2, df$Replicate)
df$Replicate = ifelse(grepl("-3-", df$Sample), 3, df$Replicate)
df$Replicate = ifelse(grepl("-4-", df$Sample), 4, df$Replicate)
table(df$Replicate)

# Primer pulse treatment

df$Primer_pulse = NA
df$Primer_pulse = ifelse(grepl("-t1-", df$Sample), 5, df$Primer_pulse)
df$Primer_pulse = ifelse(grepl("-t2-", df$Sample), 50, df$Primer_pulse)
df$Primer_pulse = ifelse(grepl("-t3-", df$Sample), 500, df$Primer_pulse)
df$Primer_pulse = ifelse(grepl("-t4-", df$Sample), 0, df$Primer_pulse)
table(df$Primer_pulse)
```

```r
# Inspect only nonsynonymous mutations

unique(df$EFFECT)

df_nonsyn = df[!df$EFFECT %in% c('intergenic_region',
                                 'synonymous_variant',
                                 'conservative_inframe_deletion',
                                 'non_coding_transcript_variant',
                                 'conservative_inframe_insertion',
                                 'gene_fusion',
                                 'transcript_ablation',
                                 'conservative_inframe_insertion&splice_region_variant'),]
dim(df)
dim(df_nonsyn) # 9297/17173 variants retained
```

```r
# Inspect variant allele frequency

df_nonsyn$COUNT = 1
```

```r
# Inspect variant allele frequency

df_nonsyn = df_nonsyn[order(df_nonsyn$AF, decreasing = T),]
ggplot(df_nonsyn, aes(x = AF)) +
    geom_histogram() +
    geom_vline(xintercept = 0.07) +
    geom_vline(xintercept = 0.5)

# Let's remove lower distribution (potential seq errors and less interesting variants)

df_highfreq = df_nonsyn[df_nonsyn$AF > 0.5,]
nrow(df_nonsyn)
nrow(df_highfreq) # 217/9297 mutations retained
```

```r
# Inspect gene targets in independent populations

df_highfreq$COUNT = 1
nonsyn_genes = aggregate(COUNT ~ Sample + GENE, data = df_highfreq, FUN = sum, na.rm = TRUE)
nonsyn_genes = nonsyn_genes[order(nonsyn_genes$COUNT, decreasing = T),]
head(nonsyn_genes)

ggplot(nonsyn_genes, aes(x = COUNT)) +
    geom_histogram(binwidth = 1) +
    geom_vline(xintercept = 2)
```

```r
# Let's allow a maximum of two mutations in the same gene in the same population
# (most genes are only mutated once in one population and more than two may imply a false positive)

nonsyn_genes_2 = unique(nonsyn_genes$GENE[nonsyn_genes$COUNT > 2])
df_highfreq2 = df_highfreq[!(df_highfreq$GENE %in% nonsyn_genes_2),]
nrow(df_highfreq)
nrow(df_highfreq2) # 208/217 mutations retained 
length(unique(df_highfreq$GENE)) # in 39 genes
```

```r
length(unique(df_highfreq2$Sample)) # 64 samples
```

```r
unique(df_highfreq2$gene_name1)

# make new gene name column based on either generic name or locus tag

df_highfreq2$gene_name = ifelse(is.na(df_highfreq2$gene_name1) == T, df_highfreq2$GENE, as.character(df_highfreq2$gene_name1))

unique(df_highfreq2$gene_name)
```

```r
# Order genes by total frequency

domgenes = aggregate(AF~gene_name, FUN = sum, data = df_highfreq2)
domgenes = domgenes[order(domgenes$AF, decreasing = T),]
domgenelist = domgenes$gene_name

df_highfreq2$gene_name = factor(df_highfreq2$gene_name, levels = domgenelist)
```

```r
mut_plot <- ggplot(data = df_highfreq2, aes(x = HAMBI, y = gene_name, fill = AF)) +
  geom_tile() +
    scale_y_discrete(limits = rev) +
    facet_grid(. ~ factor(Primer_pulse)*Evo*Replicate) +
  #scale_fill_manual(values = c("grey", "black")) +
  #scale_x_discrete(labels = scale_labels) +
    labs(fill = "Nonsynonymous allele\nfrequency (min. 0.5)", y = "Gene hit by mutation", x = "") +
        theme_classic() +
  theme(
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank(),
    panel.background = element_blank(),
    axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1, color = "black", size = 12),
    axis.title.y = element_text(face = "bold", color = "black", size = 12),
    axis.text.y = element_text(color = "black", size = 12, face = "italic"),
      legend.text = element_text(color = "black", size = 12, , colour = "black"),
      legend.title = element_text(color = "black", size = 12, colour = "black", face = "bold")#,
      #legend.position = c(0.2, 0.2)
  )

mut_plot
```

```r
ggsave("../figs/metagen_mutations.pdf", mut_plot, device = "pdf", height = 9, width = 20)
```

### Inspect common genes per species per condition

```r
head(df_highfreq2[,-c(1:25)])
```

```r
# Aggregate by count

highfreq_aggr = aggregate(COUNT ~ HAMBI*gene_name*Primer_pulse*Evo,FUN = sum, data = df_highfreq2)
max(highfreq_aggr$COUNT)
min(highfreq_aggr$COUNT)
```

```r
unique(highfreq_aggr$HAMBI)
```

```r
# Remove genes mutated only in one population

single_genes = aggregate(COUNT ~ HAMBI*gene_name,FUN = sum, data = highfreq_aggr)
single_genes = single_genes$gene_name[single_genes$COUNT<2]
single_genes

highfreq_aggr = highfreq_aggr[!(highfreq_aggr$gene_name %in% single_genes),]
```

```r
gene_names = sort(as.character(unique(highfreq_aggr$gene_name)))
gene_names2 = c(gene_names[1:3],gene_names[13:15], gene_names[4:12])
gene_names2

highfreq_aggr$gene_name = factor(highfreq_aggr$gene_name, levels = gene_names2)
```

```r
highfreq_aggr$Evo = factor(highfreq_aggr$Evo, levels = c("All ancestral", "Evolved 1972", "Evolved 1977", "All evolved"))
```

```r
highfreq_aggr$gene_name_species = paste(highfreq_aggr$HAMBI, "gene", highfreq_aggr$gene_name)
```

```r
mut_plot2 <- ggplot(data = highfreq_aggr, aes(x = factor(Primer_pulse), y = gene_name_species, fill = COUNT)) +
  geom_tile() +
scale_fill_gradient(low = "pink", high = "purple") +
    scale_y_discrete(limits = rev, position = "right") +
    facet_grid(. ~ Evo) +
  #scale_fill_manual(values = c("grey", "black")) +
  #scale_x_discrete(labels = scale_labels) +
    labs(fill = "Number of replicate populations\nhit by mutation (max. 4)", y = "", x = "Primer pulse") +
        theme_classic() +
  theme(
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank(),
    panel.background = element_blank(),
    axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1, color = "black", size = 12),
    axis.title.y = element_text(face = "bold", color = "black", size = 12),
    axis.text.y = element_text(color = "black", size = 12, face = "italic"),
      legend.text = element_text(color = "black", size = 12, , colour = "black"),
      legend.title = element_text(color = "black", size = 12, colour = "black", face = "bold")#,
      #legend.position = c(0.2, 0.2)
  )

mut_plot2
```

```r
ggsave("../figs/metagen_mutations.pdf", mut_plot2, device = "pdf", height = 9, width = 10)
```
