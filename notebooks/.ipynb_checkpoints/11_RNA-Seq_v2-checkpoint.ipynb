# Analysis of RNASeq DE data

```r
# See: 
# https://www.bioconductor.org/packages/release/bioc/vignettes/DESeq2/inst/doc/DESeq2.html
# https://www.bioconductor.org/packages/release/workflows/vignettes/rnaseqGene/inst/doc/rnaseqGene.html
# https://www.youtube.com/watch?v=ht1r34-ifVI
```

```r
list.files("../../../New_RNASeq_data/results")
```

```r
# Read in data

comms = read.table('../../../New_RNASeq_data/results/deseq_comparing_communities_norm_RNAtotals_shrink.tsv', header = T, sep = "\t")
pulses = read.table('../../../New_RNASeq_data/results/deseq_comparing_treatments_norm_RNAtotals_shrink.tsv', header = T, sep = "\t")
head(comms)
head(pulses)
```

```r
unique(comms$treatment)
unique(comms$comparison)
unique(pulses$community)
unique(pulses$comparison)
#a. ANC (all ancestral strains)
#b. E1287 (all ancestral expect evolved 1287)
#c. E1896 (all ancestral except evolved 1896)
#d. E1972 (all ancestral except evolved 1972)
#e. E1977 (all ancestral except evolved 1977)
#f. EVO (all evolved strains)
#Treatment 1. 5 ug/ml (t1)
#Treatment 2. 50 ug/ml (t2)
#Treatment 3. 500 ug/ml (t3)
#Treatment 4. (control) 0 ug/ml (t4)
```

```r
# Rename "communities" and "treatments"

comms$treatment = factor(comms$treatment, 
                         levels = c('t4','t1','t2','t3'), 
                         labels = c(0,5,50,500))
comms$comparison = factor(comms$comparison, 
                         levels = c('a_d','a_e','a_f','d_e','d_f','e_f'),
                         labels = c('ANC vs. E1972',
                                    'ANC vs. E1977',
                                    'ANC vs. EVO',
                                    'E1972 vs. E1977',
                                    'E1972 vs. EVO',
                                    'E1977 vs. EVO'))
pulses$community = factor(pulses$community, levels = c('a','d','e',"f"),
                         labels = c("ANC", "E1972", "E1977","ALL_EVO"))
pulses$comparison = factor(pulses$comparison, 
                          levels = c('t1_t2','t1_t3','t1_t4','t2_t3','t2_t4','t3_t4'),
                         labels = c('5 vs. 50','5 vs. 500','5 vs. 0','50 vs. 500','50 vs. 0','500 vs. 0'))
```

```r
# Load dependencies

#install.packages(c("reshape2", "tidyverse"))
library(reshape2)
library(tidyverse)
```

```r
#Inspect
head(comms)
head(pulses)
```

```r
str(comms)
```

```r
# Look for ARGs using CARD data

CARD = read.delim('../data/card-data/aro_index.tsv', header = T, sep = "\t", stringsAsFactors=FALSE)
head(CARD)
```

```r
# Extract CARD data containing DE gene names present in comm data
# CARD Data August 2024 Release (version 3.3.0, 2024-08-26; https://card.mcmaster.ca/download)

commsgenes = unique(na.omit(gsub("_.*", "", comms$gene_name)))
length(commsgenes)
head(commsgenes)

CARD_comms = CARD[CARD$ARO_Name %in% commsgenes,]
nrow(CARD)
nrow(CARD_comms)
CARD_comms

write.table(CARD_comms, "../data/CARD_comms.txt", quote = F, row.names = F)

pulsegenes = unique(na.omit(gsub("_.*", "", pulses$gene_name)))
length(pulsegenes)
head(pulsegenes)

CARD_pulses = CARD[CARD$ARO_Name %in% pulsegenes,]
nrow(CARD)
nrow(CARD_pulses)
CARD_pulses

write.table(CARD_pulses, "../data/CARD_pulses.txt", quote = F, row.names = F)
```

```r
comms_1972_v2 = comms_1972
comms_1972_v2$treatment2 = paste(comms_1972_v2$comparison, comms_1972_v2$treatment, sep = "_")
comms_1972_v2 = dcast(comms_1972_v2, Gene ~ treatment2, value.var = "log2FoldChange")
comms_1972_v2[is.na(comms_1972_v2)] <- 0
head(comms_1972_v2)
nrow(comms_1972_v2)
```
