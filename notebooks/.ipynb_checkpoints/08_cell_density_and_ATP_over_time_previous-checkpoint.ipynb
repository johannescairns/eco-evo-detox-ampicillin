```r
library(MASS)
library(nlme)
library(Polychrome)
library(scales)
library(tidyverse)
library(withr)
```

```r
# Read in data

df = read.table("../data/cell_density_ATP.txt", header = T, check.names = F, dec = ",")
head(df)
```

```r
unique(df$day)
```

```r
# Add treatment information

# Evo treatment

df$Evo = NA
df$Evo = ifelse(grepl("^a_", df$sample_id), "All ancestral", df$Evo)
df$Evo = ifelse(grepl("^d_", df$sample_id), "Evolved 1972", df$Evo)
df$Evo = ifelse(grepl("^e_", df$sample_id), "Evolved 1977", df$Evo)
df$Evo = ifelse(grepl("^f_", df$sample_id), "All evolved", df$Evo)
table(df$Evo)

# Replicate

df$Replicate = NA
df$Replicate = ifelse(grepl("_1_", df$sample_id), 1, df$Replicate)
df$Replicate = ifelse(grepl("_2_", df$sample_id), 2, df$Replicate)
df$Replicate = ifelse(grepl("_3_", df$sample_id), 3, df$Replicate)
df$Replicate = ifelse(grepl("_4_", df$sample_id), 4, df$Replicate)
table(df$Replicate)

# Primer pulse treatment

df$Primer_pulse = NA
df$Primer_pulse = ifelse(grepl("_t1_", df$sample_id), 5, df$Primer_pulse)
df$Primer_pulse = ifelse(grepl("_t2_", df$sample_id), 50, df$Primer_pulse)
df$Primer_pulse = ifelse(grepl("_t3_", df$sample_id), 500, df$Primer_pulse)
df$Primer_pulse = ifelse(grepl("_t4_", df$sample_id), 0, df$Primer_pulse)
table(df$Primer_pulse)
```

```r
colnames(df)
```

```r
scientific_10 = function(x) {
  ifelse(
    x==0, "0",
    parse(text = sub("e[+]?", " %*% 10^", scientific_format()(x)))
  )
} 
```

```r
df$Evo = factor(df$Evo, levels = c("All ancestral", "Evolved 1972", "Evolved 1977", "All evolved"))
```

```r
p_cells<-ggplot(data=na.omit(df), aes(x=day, y=cell_number_ml)) +
    geom_rect(aes(xmin = 4, xmax = 8, ymin = -Inf, ymax = Inf), 
              fill="red", alpha = 0.005) +
    geom_rect(aes(xmin = 12, xmax = 16, ymin = -Inf, ymax = Inf), 
              fill="red", alpha = 0.005) +
  stat_summary(fun.y = mean,
               fun.ymin = function(x) mean(x) - sd(x), 
               fun.ymax = function(x) mean(x) + sd(x), 
               geom = "pointrange") +
  stat_summary(fun.y = mean,
               geom = "line") +
    facet_grid(Evo ~ factor(Primer_pulse)) +
    labs(y = bquote("Cells mL"^-1), x = "Day") +
    scale_y_continuous(label = scientific_10) +
    scale_x_continuous(breaks = c(4,8,12, 16,20)) +
    #theme_classic() +
    theme(axis.title.x = element_text(size = 16, color = "black"),
         axis.title.y = element_text(size = 16, color = "black"))
p_cells
ggsave("../figs/cells.pdf", p_cells, device = "pdf", height = 5, width = 6)
```

```r
# Cell density stats baseline

baseline = na.omit(df[df$day %in% c(1,2,3,4),])

M1_baseline = gls(cell_number_ml ~ Evo + day, 
         data = baseline, 
         #weights = varIdent(form = ~1 | Streptomycin),
         correlation = corAR1(form = ~ 1 | sample),
         method = "ML")

M2_baseline = stepAIC(M1_baseline, 
             scope = list(upper = ~Evo * day, lower = ~1), 
             trace = FALSE)

anova(M2_baseline)

# sig. effect of pre-exposure

# pairwise contrasts

comm.emm.s = emmeans(M2_baseline, pairwise ~ Evo)
pairs(comm.emm.s) 
# all anc and evo 1977 have lower abundance compared to all evo and evo 1972
```

```r
# Cell density stats primer pulse

primer = na.omit(df[df$day %in% c(5,6,7,8),])
primer$Primer_pulse = factor(primer$Primer_pulse, levels = c(0, 5, 50, 500),
                            labels = c("No", "Low", "Medium", "High"))

M1_primer = gls(cell_number_ml ~ Evo + Primer_pulse + day, 
         data = primer, 
         #weights = varIdent(form = ~1 | Streptomycin),
         correlation = corAR1(form = ~ 1 | sample),
         method = "ML")

M2_primer = stepAIC(M1_primer, 
             scope = list(upper = ~Evo * Primer_pulse * day, lower = ~1), 
             trace = FALSE)

anova(M2_primer)

# sig. effect of pre-exposure, primer pulse and day (increase over time)

# pairwise contrasts

comm.emm.s = emmeans(M2_primer, pairwise ~ Evo)
pairs(comm.emm.s) 
# all anc and evo 1977 have lower abundance compared to all evo and evo 1972

comm.emm.s2 = emmeans(M2_primer, pairwise ~ Primer_pulse)
pairs(comm.emm.s2) 
```

```r
# Cell density stats main pulse

main = na.omit(df[df$day %in% c(13,14,15,16),])
main$Primer_pulse = factor(main$Primer_pulse, levels = c(0, 5, 50, 500),
                            labels = c("No", "Low", "Medium", "High"))

M1_main = gls(cell_number_ml ~ Evo + Primer_pulse + day, 
         data = main, 
         #weights = varIdent(form = ~1 | Streptomycin),
         correlation = corAR1(form = ~ 1 | sample),
         method = "ML")

M2_main = stepAIC(M1_main, 
             scope = list(upper = ~Evo * Primer_pulse * day, lower = ~1), 
             trace = FALSE)

anova(M2_main)

# sig. effect of pre-exposure, main pulse and day (increase over time)

# pairwise contrasts

comm.emm.s = emmeans(M2_main, pairwise ~ Evo)
pairs(comm.emm.s) 
# all anc and evo 1977 have lower abundance compared to all evo and evo 1972

comm.emm.s2 = emmeans(M2_main, pairwise ~ Primer_pulse)
pairs(comm.emm.s2) 
```
