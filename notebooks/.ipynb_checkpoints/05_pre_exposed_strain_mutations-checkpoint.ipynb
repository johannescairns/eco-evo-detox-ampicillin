```r
library(tidyverse)
library(here)
library(withr)
library(Polychrome)
library(rcartocolor)
library(scales)
```

```r
df = read_tsv("../data/amp_mutation_sp_summary.tsv")
head(df)
```

```r
# Extract species without and with mutations

df_nomut = df[df$mutation_detected == "no",]
df_nomut$species # 11/23 species have no SNPs or indels

df_mut = df[df$mutation_detected == "yes",]

# Remove unannotated mutations (not affecting coding regions)

nrow(df_mut)
df_mut = df_mut[!is.na(df_mut$gene_locus),]
nrow(df_mut)
head(df_mut) # 58 mutations affecting coding regions

unique(df_mut$species) # 11/23 species have SNPs or indels in coding regions
```

```r
# Add count data for plotting (no allele freq, since isogenic populations)

df_mut$count = 1

# Check number of mutations per species

aggregate(count ~ species, FUN = sum, data = df_mut)
```

```r
scale_labels = c(expression(paste(italic('Pseudomonas putida'), ' HAMBI 6')),                 
                 expression(paste(italic('Agrobacterium tumefaciens'), ' HAMBI 105')),
                 expression(paste(italic('Sphingobium yanoikuyae'), ' HAMBI 1842')),
                 expression(paste(italic('Sphingobacterium spiritivorum') ,' HAMBI 1896')),
                 expression(paste(italic('Aeromonas caviae'), ' HAMBI 1972')),
                 expression(paste(italic('Pseudomonas chlororaphis'), ' HAMBI 1977')),
                 expression(paste(italic('Chitinophaga sancti'), ' HAMBI 1988')),
                 expression(paste(italic('Bordetella avium'), ' HAMBI 2160')),
                 expression(paste(italic('Paraburkholderia kururiensis') ,' HAMBI 2494')),
                 expression(paste(italic('Moraxella canis'), ' HAMBI 2792')),
                 expression(paste(italic('Microvirga lotononidis'), ' HAMBI 3237')))
scale_labels
```

```r
# Add simplified effect categories and add gene code where generic gene name missing

#unique(df_mut$effect)
df_mut$effect2 = ifelse(grepl("synonymous", df_mut$effect), "Synonymous", "Nonsynonymous")
df_mut$gene2 = ifelse(is.na(df_mut$gene_name) == TRUE, df_mut$gene_locus, df_mut$gene_name)
head(df_mut)
```

```r
# Check for gene recurrence

df_rec_within_species = aggregate(count ~ gene2*species, FUN = sum, data = df_mut)
df_rec_within_species$count = 1
df_rec_between_species = aggregate(count ~ gene2, FUN = sum, data = df_rec_within_species)
df_rec_between_species = df_rec_between_species[order(df_rec_between_species$count,
                                                     decreasing = T),]
head(df_rec_between_species)
# no gene is recurrently mutated in independent populations
```

```r
x = sort(unique(df_mut$gene2))
gene_names = c(x[1:5], x[26:32], x[6:25])
gene_names

df_mut$gene2 = factor(df_mut$gene2, levels = gene_names)
```

```r
df_mut$effect2 = factor(df_mut$effect2, levels = c("Synonymous", "Nonsynonymous"))
```

```r
mut_plot <- ggplot(data = df_mut, aes(x = species, y = gene2, fill = effect2)) +
  geom_tile() +
    scale_y_discrete(limits = rev) +
  scale_fill_manual(values = c("grey", "black")) +
  scale_x_discrete(labels = scale_labels) +
    labs(fill = "Mutation type", y = "Gene hit by mutation", x = "") +
        theme_classic() +
  theme(
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank(),
    panel.background = element_blank(),
    axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1, color = "black", size = 12),
    axis.title.y = element_text(face = "bold", color = "black", size = 12),
    axis.text.y = element_text(color = "black", size = 12, face = "italic"),
      legend.text = element_text(color = "black", size = 12, , colour = "black"),
      legend.title = element_text(color = "black", size = 12, colour = "black", face = "bold")#,
      #legend.position = c(0.2, 0.2)
  )

mut_plot
```

```r
ggsave("../figs/mutations.pdf", mut_plot, device = "pdf", height = 9, width = 8)
```
