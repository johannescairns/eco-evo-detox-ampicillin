```r
library(tidyverse)
library(here)
library(scales)
library(stringi)
```

```r
df = read.table("../data/AMP_growth_k_auc.txt", header = T, check.names = F, sep = "\t")
head(df)
```

```r
# Use empirical area under the curve which is not subject to parameter estimation errors

p1 = ggplot(df[df$AB == 0,], aes(factor(Species), auc_e, colour = Pop)) + 
  stat_summary(fun.data = "mean_cl_boot", 
               position = position_dodge(width=0.75)) + 
  geom_point(alpha = 0.35, 
             position = position_jitterdodge(seed=123, jitter.width=0, dodge.width = 0.75)) +
  #scale_y_continuous(trans = "log1p", breaks=c(unique(round(rev(my_split[[1]]$ampicillin))))) +
  #scale_color_carto_d(palette="Safe") +
  #labs(x="", y="Ampicillin IC50 (ug/ml)") +
  theme_bw() + 
  theme(
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank(),
    panel.background = element_blank(),
    axis.text.x = element_text(angle = 45, vjust = 1, hjust=1)
  )
p1
```

```r
# Areas under the curve do not correspond to strain competitive fitness
# Therefore, let's extract strain-specific information about fitness of evo vs. anc strain

df_means = aggregate(auc_e ~ Pop*Species, FUN = mean, data = df[df$AB == 0 &
                                                                df$Species != "only medium",])
head(df_means)
df_means_anc = df_means[df_means$Pop == "ANC",-1]
colnames(df_means_anc) = c("Species", "AUC_E_ANC")
df_means_evo = df_means[df_means$Pop == "EVO",-1]
colnames(df_means_evo) = c("Species", "AUC_E_EVO")
df_means2 = merge(df_means_anc, df_means_evo, by = "Species")
df_means2$W_EVO_VS_ANC = df_means2$AUC_E_EVO / df_means2$AUC_E_ANC
df_means2$Species = as.numeric(as.character(df_means2$Species))
df_means2 = df_means2[order(df_means2$Species),]
head(df_means2)
```

```r
df_means2$Species = factor(df_means2$Species, levels = df_means2$Species,
                          labels = paste("HAMBI", df_means2$Species, sep = "_"))
```

```r
df_means2$Treat = ifelse(grepl("HAMBI_1972", df_means2$Species) | grepl("HAMBI_1977", df_means2$Species), "Single species pre-evolution treatment", "Others")
```

```r
p2 = ggplot(df_means2,
            aes(Species, W_EVO_VS_ANC, colour = Treat)) + 
  geom_point(size = 3) +
    scale_colour_manual(values = c("grey", "red")) +
    geom_hline(yintercept = 1, linetype = 2) +
    scale_y_log10(breaks = c(0,0.2,0.4,0.6,0.8,1,1.2,1.4,1.6, 10)) +
    labs(y = "Relative fitness of evolved strain\n(area under growth curve relative to ancestral strain)",
        x = "") +
  theme_classic() + 
  theme(
    axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1, size = 14, colour = "black"),
    axis.text.y = element_text(size = 14, colour = "black"),
      axis.title = element_text(size = 14, colour = "black"),
      legend.title = element_blank(),
      legend.text = element_text(size = 14, colour = "black")
    
  )
p2
```

```r
# Save plot

ggsave("../figs/Evo_strain_fitness.pdf", p2, width=12, height=9, units="in", device="pdf")
```

```r
# Read in species abundance data

species = read.table("../data/counts_final_fmt.txt", header = T, check.names = F)
head(species)
```

```r
scale_labels = c(expression(paste(italic('Pseudomonas putida'), ' HAMBI 6')),                 
                 expression(paste(italic('Acinetobacter lwoffii'), ' HAMBI 97')),
                 expression(paste(italic('Agrobacterium tumefaciens'), ' HAMBI 105')),
                 expression(paste(italic('Brevundimonas bullata'), ' HAMBI 262')),
                 expression(paste(italic('Comamonas testosteroni'), ' HAMBI 403')),
                 expression(paste(italic('Hafnia alvei'), ' HAMBI 1279')),
                 expression(paste(italic('Citrobacter koseri'), ' HAMBI 1287')),
                 expression(paste(italic('Morganella morganii'), ' HAMBI 1292')),
                 expression(paste(italic('Kluyvera intermedia'), ' HAMBI 1299')),
                 expression(paste(italic('Sphingobium yanoikuyae'), ' HAMBI 1842')),
                 expression(paste(italic('Sphingobacterium spiritivorum') ,' HAMBI 1896')),
                 expression(paste(italic('Aeromonas caviae'), ' HAMBI 1972')),
                 expression(paste(italic('Pseudomonas chlororaphis'), ' HAMBI 1977')),
                 expression(paste(italic('Chitinophaga sancti'), ' HAMBI 1988')),
                 expression(paste(italic('Paraburkholderia caryophylli'), ' HAMBI 2159')),
                 expression(paste(italic('Bordetella avium'), ' HAMBI 2160')),
                 expression(paste(italic('Cupriavidus necator'), ' HAMBI 2164')),
                 expression(paste(italic('Paracoccus denitrificans'), ' HAMBI 2443')),
                 expression(paste(italic('Paraburkholderia kururiensis') ,' HAMBI 2494')),
                 expression(paste(italic('Stenotrophomonas maltophilia'), ' HAMBI 2659')),
                 expression(paste(italic('Moraxella canis'), ' HAMBI 2792')),
                 expression(paste(italic('Niabella yanshanensis'), ' HAMBI 3031')),
                 expression(paste(italic('Microvirga lotononidis'), ' HAMBI 3237')))
scale_labels
```

```r
# Add treatment information

# Day

species$Day = NA
species$Day = ifelse(grepl("d4", species$sample), 4, species$Day)
species$Day = ifelse(grepl("d8", species$sample), 8, species$Day)
species$Day = ifelse(grepl("d12", species$sample), 12, species$Day)
species$Day = ifelse(grepl("d16", species$sample), 16, species$Day)
species$Day = ifelse(grepl("d20", species$sample), 20, species$Day)
table(species$Day)

# Evo treatment

species$Evo = NA
species$Evo = ifelse(grepl("^a_", species$sample), "All ancestral", species$Evo)
species$Evo = ifelse(grepl("^d_", species$sample), "Evolved 1972", species$Evo)
species$Evo = ifelse(grepl("^e_", species$sample), "Evolved 1977", species$Evo)
species$Evo = ifelse(grepl("^f_", species$sample), "All evolved", species$Evo)
table(species$Evo)

# Replicate

species$Replicate = NA
species$Replicate = ifelse(grepl("_1_", species$sample), 1, species$Replicate)
species$Replicate = ifelse(grepl("_2_", species$sample), 2, species$Replicate)
species$Replicate = ifelse(grepl("_3_", species$sample), 3, species$Replicate)
species$Replicate = ifelse(grepl("_4_", species$sample), 4, species$Replicate)
table(species$Replicate)

# Primer pulse treatment

species$Primer_pulse = NA
species$Primer_pulse = ifelse(grepl("_t1_", species$sample), 5, species$Primer_pulse)
species$Primer_pulse = ifelse(grepl("_t2_", species$sample), 50, species$Primer_pulse)
species$Primer_pulse = ifelse(grepl("_t3_", species$sample), 500, species$Primer_pulse)
species$Primer_pulse = ifelse(grepl("_t4_", species$sample), 0, species$Primer_pulse)
table(species$Primer_pulse)
```

```r
# Remove negative controls

nrow(species)
species = species[!grepl("control", species$sample),]
nrow(species)
```

```r
# Check that equal amount of data for all treatments

table(species$Day)
table(species$Evo)
table(species$Replicate)
table(species$Primer_pulse)

# ok
```

```r
species$Evo = factor(species$Evo, levels = c("All ancestral", "Evolved 1972", "Evolved 1977", "All evolved"))
```

```r
head(species)
```

```r
# Aggregate species abundance at day 4 after strain establishment and prior to pre-pulse

# Perform separately for all anc and all evo

all_anc_comp = aggregate(data = species[species$Evo == "All ancestral" &
                                        species$Day == 4,], 
                        f_correct ~ strainID, FUN = mean)
colnames(all_anc_comp) = c("strainID", "f_all_anc")
all_evo_comp = aggregate(data = species[species$Evo == "All evolved" &
                                        species$Day == 4,], 
                        f_correct ~ strainID, FUN = mean)
colnames(all_evo_comp) = c("strainID", "f_all_evo")
day4 = merge(all_anc_comp, all_evo_comp, by = "strainID")
```

```r
# Remove strains absent from either community

nrow(day4)
day4 = day4[!(day4$f_all_anc == 0 | day4$f_all_evo == 0 ),]
nrow(day4)
```

```r
# Add relative change evo. vs anc

day4$f_all_evo_vs_all_anc = day4$f_all_evo/day4$f_all_anc
```

```r
df_means2$Species = gsub("HAMBI_6", "HAMBI_0006", df_means2$Species)
df_means2$Species = gsub("HAMBI_97", "HAMBI_0097", df_means2$Species)
df_means2$Species = gsub("HAMBI_105", "HAMBI_0105", df_means2$Species)
df_means2$Species = gsub("HAMBI_262", "HAMBI_0262", df_means2$Species)
df_means2$Species = gsub("HAMBI_403", "HAMBI_0403", df_means2$Species)
unique(df_means2$Species)
```

```r
colnames(df_means2)[1] = "strainID"
nrow(df_means2)
df_means2 = df_means2[df_means2$strainID %in% unique(day4$strainID),]
nrow(df_means2)
head(df_means2)
```

```r
df2 = merge(day4, df_means2, by = "strainID")
nrow(day4)
nrow(df_means2)
nrow(df2)
head(df2)
```

```r
# Aggregate species abundance in control condition (no pre-pulse,
# no pre-evolution) at days 12 and 16 (before and after main pulse)

# Perform separately for all anc and all evo

all_anc_12 = aggregate(data = species[species$Evo == "All ancestral" &
                                        species$Day == 12 & species$Primer_pulse == 0,], 
                        f_correct ~ strainID, FUN = mean)
colnames(all_anc_12) = c("strainID", "f_d12")
all_anc_16 = aggregate(data = species[species$Evo == "All ancestral" &
                                        species$Day == 16 & species$Primer_pulse == 0,], 
                        f_correct ~ strainID, FUN = mean)
colnames(all_anc_16) = c("strainID", "f_d16")
all_anc_12_vs_16 = merge(all_anc_12, all_anc_16, by = "strainID")
```

```r
# Remove strains absent both communities

nrow(all_anc_12_vs_16)
all_anc_12_vs_16 = all_anc_12_vs_16[!(all_anc_12_vs_16$f_d12 == 0 & all_anc_12_vs_16$f_d16 == 0 ),]
nrow(all_anc_12_vs_16) # 15 left
```

```r
# Add pseudocount for cases where strain is at zero abundance

min_val = min(c(all_anc_12_vs_16$f_d12[all_anc_12_vs_16$f_d12>0], all_anc_12_vs_16$f_d16[all_anc_12_vs_16$f_d16>0]))
min_val

all_anc_12_vs_16$f_d12 = ifelse(all_anc_12_vs_16$f_d12 == 0, min_val, all_anc_12_vs_16$f_d12)
all_anc_12_vs_16$f_d16 = ifelse(all_anc_12_vs_16$f_d16 == 0, min_val, all_anc_12_vs_16$f_d16)

head(all_anc_12_vs_16)
```

```r
# Add relative change evo. vs anc

all_anc_12_vs_16$f_16_vs_12 = all_anc_12_vs_16$f_d16/all_anc_12_vs_16$f_d12

head(all_anc_12_vs_16)
```

```r
library(reshape2)
all_anc_comp = melt(all_anc_12_vs_16[,1:3])
colnames(all_anc_comp) = c("Species", "Day", "Frequency")
all_anc_comp$Day = ifelse(all_anc_comp$Day == "f_d12", 12, 16)
#all_anc_comp
```

```r
unique(all_anc_comp$Species)
```

```r
all_anc_12_vs_16 = all_anc_12_vs_16[order(all_anc_12_vs_16$f_16_vs_12, decreasing = T),]
head(all_anc_12_vs_16)
all_anc_12_vs_16$strainID
```
