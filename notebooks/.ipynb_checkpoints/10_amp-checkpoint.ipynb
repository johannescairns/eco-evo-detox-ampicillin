# Analysis of AMP data

```r
# Read in data

df = read.table("../data/AMP_conc.txt", header = T, sep = "\t")
head(df)
```

```r
unique(df$Evo)
```

```r
str(df)
```

```r
#+++++++++++++++++++++++++
# Function to calculate the mean and the standard deviation
  # for each group
#+++++++++++++++++++++++++
# data : a data frame
# varname : the name of a column containing the variable
  #to be summariezed
# groupnames : vector of column names to be used as
  # grouping variables
data_summary <- function(data, varname, groupnames){
  require(plyr)
  summary_func <- function(x, col){
    c(mean = mean(x[[col]], na.rm=TRUE),
      sd = sd(x[[col]], na.rm=TRUE))
  }
  data_sum<-ddply(data, groupnames, .fun=summary_func,
                  varname)
  data_sum <- rename(data_sum, c("mean" = varname))
 return(data_sum)
}
```

```r
df2 = data_summary(df, varname="AMP_conc", 
                    groupnames=c("Evo", "AB", "Day"))
head(df2)
```

```r
# Load dependencies

library(reshape2)
library(tidyverse)
```

```r
library(RColorBrewer)
violets <- colorRampPalette(c("white","violet"))
violets(5)
```

```r
unique(df2$Evo)
```

```r
df2$Evo = factor(df2$Evo, 
                 levels = c("ANC", "E1972", "E1977", "EVO", "No_BAC"),
                 labels = c("All ancestral", "Evolved 1972", "Evolved 1977", "All evolved", "Bacteria-free control"))
```

```r
# Visualize

p<- ggplot(df2, aes(x=Day, y=AMP_conc, group=factor(AB), color=factor(AB))) + 
 geom_rect(aes(xmin=4, xmax=8,ymin=0,ymax=5), fill='#FADFFA', color = NA)+
geom_rect(aes(xmin=4, xmax=8,ymin=5,ymax=50), fill='#F6C0F6', color = NA)+
geom_rect(aes(xmin=4, xmax=8,ymin=50,ymax=500), fill='#EE82EE', color = NA)+
    geom_rect(aes(xmin=12, xmax=16,ymin=0,ymax=500), fill='#EE82EE',color = NA)+
  #geom_errorbar(aes(ymin=AMP_conc-sd, ymax=AMP_conc+sd), width=.3, color = "black") +
  #geom_errorbar(aes(ymin=AMP_conc-sd, ymax=AMP_conc+sd), width=.2) +
geom_pointrange(aes(ymin=AMP_conc-sd, ymax=AMP_conc+sd), size = 0*1.5, color = "black") +
geom_pointrange(aes(ymin=AMP_conc-sd, ymax=AMP_conc+sd)) +
geom_line(size=1*1.1, aes(group= factor(AB)), colour = "black")+
  geom_line() +
geom_point(pch=21, colour= "black", aes(fill =factor(AB)), size = 2.5)+
    scale_color_manual(values = c('grey','#FADFFA','#F6C0F6','#EE82EE')) +
scale_fill_manual(values = c('grey','#FADFFA','#F6C0F6','#EE82EE')) +

    facet_wrap(~Evo, ncol = 1) +
    scale_y_log10(breaks = c(0,5,50,500)) +
    scale_x_continuous(breaks = c(0, 4, 8, 12,16,20)) +
    theme_classic() +
    labs(color = "Pre-pulse ampicillin\nconcentration\n(main pulse 500 for all)", y = "Ampicillin concentration") +
guides(fill = FALSE, size = FALSE)
p
```

```r
ggsave("../figs/amp.pdf", width = 6, height = 8)
```

```r
# Stats
```

```r
library(MASS)
library(nlme)
```

```r
head(df)
```

```r
M1 = gls(AMP_conc ~ factor(AB) + Evo + Day, data = df, correlation = corAR1(form = ~ 1 | Sample), method = "ML") 
# use stepwise model selection to find best model 
M2 = stepAIC(M1, scope = list(upper = ~factor(AB) * Evo * Day, lower = ~1), trace = FALSE)
```

```r
summary(M2)
```

```r
anova(M2)
```

```r
# Posthoc test

library(emmeans)
```

```r
comm.emm.s = emmeans(M2, pairwise ~ Evo) 
pairs(comm.emm.s)

# No bac differs from all else
```

```r
# Remove no bac and retest

df_bac = df[!df$Evo %in% "No_BAC",]
head(df_bac)
```

```r
unique(df_bac$Evo)
```

```r
M1 = gls(AMP_conc ~ factor(AB) + Evo + factor(Day), data = df_bac, correlation = corAR1(form = ~ 1 | Sample), method = "ML") 
# use stepwise model selection to find best model 
M2 = stepAIC(M1, scope = list(upper = ~factor(AB) * Evo * factor(Day), lower = ~1), trace = FALSE)
```

```r
anova(M2)
```

```r
comm.emm.s = emmeans(M2, pairwise ~ factor(AB) | factor(Day)) 
pairs(comm.emm.s)
```

```r
head(df_bac)
```

```r
# Summarize over evo

p2 = ggplot(df_bac, aes(Day, AMP_conc, group = factor(AB), color = factor(AB))) + 
    geom_point() +
    stat_summary(fun.data = "mean_cl_boot", linewidth = 2, size = 2, alpha = 0.5) +
    theme_classic() +
    scale_x_continuous(breaks = c(8,12,16,20)) +
scale_color_manual(values = c('grey','#FADFFA','#F6C0F6','#EE82EE')) +
    labs(y = "Ampicillin concentration", color = "Pre-pulse ampicillin\nconcentration")
p2
```

```r
#+++++++++++++++++++++++++
# Function to calculate the mean and the standard deviation
  # for each group
#+++++++++++++++++++++++++
# data : a data frame
# varname : the name of a column containing the variable
  #to be summariezed
# groupnames : vector of column names to be used as
  # grouping variables
data_summary <- function(data, varname, groupnames){
  require(plyr)
  summary_func <- function(x, col){
    c(mean = mean(x[[col]], na.rm=TRUE),
      sd = sd(x[[col]], na.rm=TRUE))
  }
  data_sum<-ddply(data, groupnames, .fun=summary_func,
                  varname)
  data_sum <- rename(data_sum, c("mean" = varname))
 return(data_sum)
}
```

```r
df$greater_than_max = ifelse(df$AMP_conc > zero_max, 1, 0)
head(df)
```

```r
df$Treat = paste(df$Evo, df$AB, df$Day, sep = "_")
head(df)
```

```r
df_Treat = aggregate(data = df, greater_than_max ~ Treat, FUN = sum)
df_Treat[df_Treat$greater_than_max==1,]
```
