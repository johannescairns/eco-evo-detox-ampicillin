```r
# Combine species abundance data into single data frame
```

```r
# Read in strain_table.txt data

st = list.files(pattern = "strain_table.txt$", recursive = TRUE)
stnames = gsub("output/", "", st)
stnames = gsub("/strain_table.txt", "", stnames)
st_list = lapply(st, read.table,header = T, check.names = F)
names(st_list) = stnames
head(st_list[[1]])
```

```r
# Add sample id, freq and normalization data

st_list2 = list()
for(i in 1:length(st_list)){
    a = st_list[[i]]
    a$SAMPLE = names(st_list[i])
    colnames(a) = c("SPECIES", "ABUNDANCE", "SAMPLE")
    a$SAMPLE = gsub("_S.*", "", a$SAMPLE)
    TOT = sum(a$ABUNDANCE)
    a$F_ABUNDANCE = a$ABUNDANCE/TOT
    st_list2[[i]] = a
}
names(st_list2)=stnames

head(st_list2[[1]])
```

```r
# Read in strain_table_normalized.txt data

stn = list.files(pattern = "strain_table_normalized.txt$", recursive = TRUE)
stnnames = gsub("output/", "", stn)
stnnames = gsub("/strain_table_normalized.txt", "", stnnames)
stn_list = lapply(stn, read.table,header = T, check.names = F)
names(stn_list) = stnames
head(stn_list[[1]])
```

```r
# Add sample id, freq and normalization data

stn_list2 = list()
for(i in 1:length(stn_list)){
    a = stn_list[[i]]
    a$SAMPLE = names(stn_list[i])
    colnames(a) = c("SPECIES", "ABUNDANCE_NORM", "SAMPLE")
    a$SAMPLE = gsub("_S.*", "", a$SAMPLE)
    TOT = sum(a$ABUNDANCE_NORM)
    a$F_ABUNDANCE_NORM = a$ABUNDANCE_NORM/TOT
    stn_list2[[i]] = a
}
names(stn_list2)=stnnames

head(stn_list2[[1]])
```

```r
# Concatenate lists into data frames

stdf=Reduce(function(...) merge(..., all=T), st_list2)
stndf=Reduce(function(...) merge(..., all=T), stn_list2)
head(stdf)
head(stndf)
```

```r
# Merge

nrow(stdf)
nrow(stndf)
species_table = merge(stdf, stndf, by = c("SPECIES", "SAMPLE"))
nrow(species_table)
head(species_table)
```

```r
species_table = species_table[order(species_table$SAMPLE, species_table$SPECIES),]
nrow(species_table)
head(species_table)
```

```r
length(unique(species_table$SAMPLE))
```

```r
# Write out

write.table(species_table, row.names = FALSE, "species_table.txt", sep = "\t", quote = F)
```
